```python
%matplotlib notebook
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
from sympy import MatMul
```

```python
t,x,y,z,r,o=sp.symbols('t,x,y,z,r,o')
a=5
#u0=4*np.pi*10**-7
v1=1
```

```python
Cil=sp.Matrix([[sp.cos(o),sp.sin(o),0],[-sp.sin(o),sp.cos(o),0],[0,0,1]])
#MatMul(Cil,rp)
Cart=sp.Matrix([[sp.cos(o),-sp.sin(o),0],[sp.sin(o),sp.cos(o),0],[0,0,1]])
```

```python
r1=sp.Matrix([r,0,z])#Vectro de posicion cilindrico arbitrario
rp=sp.Matrix([a*sp.cos(v1*t/a),a*sp.sin(v1*t/a),0]) #Posicion en cartesianas de la particula
rpc=Cil*rp #Posicion en cartesianas

vrp=sp.diff(rpc,t)
R=r1-rpc

Bc=vrp.cross(R)/R.norm()**3
Bxyz=Cart*Bc
Bx=sp.lambdify([r,o,z,t],Bxyz[0])
By=sp.lambdify([r,o,z,t],Bxyz[1])
Bz=sp.lambdify([r,o,z,t],Bxyz[2])

rr=np.linspace(0,a*2,10)
phi=np.linspace(0,2*np.pi,18)
zz=np.linspace(-4,4,8)

R,P,Z=np.meshgrid(rr,phi,zz)
X=R*np.cos(P)
Y=R*np.sin(P)
ts=0
Bxx=Bx(R,P,Z,ts)
Byy=By(R,P,Z,ts)
Bzz=Bz(R,P,Z,ts)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

BB=np.sqrt(Bxx**2+Byy**2+Bzz**2)
tt=np.linspace(0,2*np.pi,100)
ax.scatter(a*np.cos(v1*ts/a),a*np.sin(v1*ts/a),0,color='blue')
ax.plot(a*np.cos(tt),a*np.sin(tt),0,color='red')
#################################################Campo vectorial###############################
ax.quiver(X,Y,Z,Bxx,Byy,Bzz,color='black')
#ax.quiver(X,Y,Z,Bxx/BB,Byy/BB,Bzz/BB,color='black')
#ax.quiver(zz*0,zz*0,zz,Bxx*0/BB,Byy*0/BB,Bz(0,0,zz,0)/Bz(0,0,zz,0),color='black')
plt.axis('equal')
plt.show()
```

```python
fig2=plt.figure()
th=np.linspace(0,100,1000)
plt.plot(th,np.sqrt(Bx(2*a,0,0,th)**2+By(2*a,0,0,th)**2+Bz(2*a,0,0,th)**2),color='black',label='|B|')
#plt.plot(th,Bx(2*a,0,0,th),label='Bx')
#plt.plot(th,By(2*a,0,0,th),label='By')
plt.plot(th,Bz(2*a,0,0,th),label='Bz')
plt.grid('on')
plt.legend()
plt.title('Campo magnetico B(2a,0,0) vs tiempo')
plt.ylabel('B(2a,0,0)')
plt.xlabel('Tiempo(s)')
#plt.axis('equal')
```

```python
fig3=plt.figure()
plt.plot(tt,Bz(0,0,0,tt),label='Bz(r=0,phi=0,z=0) vs t')
plt.plot(tt,Bz(0,0,2,tt),label='Bz(r=0,phi=0,z=2) vs t')
plt.plot(tt,Bz(0,0,10,tt),label='Bz(r=0,phi=0,z=10) vs t')
plt.plot(tt,Bz(0,0,-5,tt),label='Bz(r=0,phi=0,z=-5) vs t')
plt.grid('on')
plt.title(' Bz en  eje z vs tiempo(s)')
plt.legend()
```

```python
q1=1
q2=2*q1
v2=3*v1

r2=sp.Matrix([0,0,-v2*t])
vr2=sp.diff(r2,t)

F=q2*(vr2.cross(Bc))
F.subs(o,0).subs(r,0).subs(z,-v2*t)
```
